## PCA Visualization

```python
# Performing Normalization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
print(scaled_data)
```

```python
scaled_df = pd.DataFrame(scaled_data, columns=cancer['feature_names'])
scaled_df.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)  # converting 30 features into 2 features (NOTE we can give any number for the component)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
scaled_df.head(10)
```

```python
x_pca.shape
```

```python
x_pca_df = pd.DataFrame(x_pca, columns=['PC1', 'PC2'])
x_pca_df.head(10)
```

```python
# We have converted the 30 features into 2 features
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(x_pca[:,0], x_pca[:,1], c=cancer['target'], cmap='plasma')
plt.xlabel('PC1')
plt.ylabel('PC2')
```
